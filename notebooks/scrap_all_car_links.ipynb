```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import selenium.webdriver.support.ui as UI
import pandas as pd 
```

```python
driver = webdriver.Chrome()
driver.get("https://www.autoscout24.com/")
```

```python
nameCarVal, nameCar = [], []
select = UI.Select(driver.find_element("id", "make"))
for option in select.options:
    val = option.get_attribute('value')
    if val != "":
        nameCarVal.append(val)  
        nameCar.append(option.text)


yearfrom = [1930, 1940, 1950, 1960, 1970, 1975, 1980, 1981, 1982,
            1983, 1984, 1985, 1986, 1987, 1988, 1989, 1990, 1991, 1992, 1993, 1994,
            1995, 1996, 1997, 1998, 1999, 2000
]
priceto = [1000, 2000, 3000, 4000, 5000, 7500, 10000, 12500, 15000,
            17500, 20000, 22500, 25000, 30000, 40000, 45000, 50000,
            60000, 70000, 80000, 90000, 100000, 150000, 200000, 300000]
```

```python
print(len(yearfrom))
print(len(priceto))
print(len(nameCar))
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())

car_links = []

file1 = open("car_links.txt", "a") 

for year in yearfrom[0:10]:
    for price in priceto:
        for name in nameCar:
            print(year, ", ", price, ", ", name)
            search_url=f"https://www.autoscout24.com/lst/{name}?sort=standard&desc=0&ustate=N,U&atype=C&priceto={price}&fregfrom={year}"      
            driver.get(search_url.format())
            elems = driver.find_elements(By.CSS_SELECTOR, "div.ListItem_wrapper__J_a_C > div.ListItem_header__uPzec > a")
            for elem in elems:
                car_ind_link = elem.get_attribute("href")
                file1.write(car_ind_link + "\n")
                car_links.append(car_ind_link)   
file1.close()

```

```python
dict = {'car_links': car_links}  
       
links_df = pd.DataFrame(dict) 
links_df.to_csv('car_links.csv') 
```
